```python
import os
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

 
# each file is organized by the variables that are changing and fixed
# the varying variables are listed first and we give the min and max values
# the values used in the grid are linearly spaced between the min and the max 
# the fixed variables are listed with the value at which they are fixed

# the user need to change the path to where they have the fits files 
path= "/Users/sarahkubiak/Desktop/WD1856/Fitsfiles/"
# load the file
hdu_list = fits.open(path+'massperiodcircularorbitfixedinclination80.fits')
hdu_list[1].header
```

```python
#Then run one of the blocks that replicated the graphs 
# these are organized by which variables are varyinging and which fixed variable is tested

#For files with varying masses and periods with a fixed inclination

a = hdu_list[1].data
#print(f["a"])
massestotest = np.linspace(hdu_list[1].header["axis1min"],hdu_list[1].header["axis1max"],a.shape[1])
periodstotest = np.linspace(hdu_list[1].header["axis0min"],hdu_list[1].header["axis0max"],a.shape[0])
meananomolytotest = np.linspace(hdu_list[1].header["axis2min"],hdu_list[1].header["axis2max"],a.shape[2])
noplanetlikihood20=hdu_list[1].header["nopl"]

print(noplanetlikihood20)
planetlikihood20 = np.exp(-1*np.min(a, axis= 2))

fig,ax= plt.subplots()
im=ax.imshow(np.maximum(np.log10(planetlikihood20/noplanetlikihood20), -4 ),vmin=-4, vmax=3, extent = [min(massestotest), max(massestotest),max(periodstotest),min(periodstotest)], aspect='auto')

fig.colorbar(im, label = 'log base ten likelihood ratio')
ax.set_ylabel(hdu_list[1].header["axis0"] + " in " + hdu_list[1].header["unit0"])
ax.set_xlabel(hdu_list[1].header["axis1"] + " in " + hdu_list[1].header["unit1"])
```
